```python
import matplotlib 
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np 
import seaborn as sns
sns.set(color_codes=True)
sns.set_style("white") 
```

```python
np.random.uniform(low=0.0, high=1.0)  
```

```python
np.random.seed(0)   # to fix the value to avoid heisunbug
def flip(num = 1):
    flips = []     
    for i in range(num): 
        num = np.random.uniform(low=0.0, high=1.0)      
        if num > 0.75: 
            flips.append('H')             
        else: 
            flips.append('T')
    return flips

flips = flip(10) 
values, counts = np.unique(flips, return_counts=True)
print(flips)
print(values, counts)
```

```python
from collections import Counter, defaultdict 
def get_freqs(flips): 
    keys = Counter(flips).keys()
    vals = Counter(flips).values()  
    return defaultdict(int, dict(zip(keys, vals)))
```

```python
freqs = get_freqs(flips)
prob_h = freqs['H'] / len(flips)
print(prob_h)
print(freqs)
```

```python
maximum_flips = 1000
probs = [] 
for num_flips in range(1, maximum_flips):
    flips = flip(num_flips)
    freqs = get_freqs(flips)
    prob_h = freqs['H'] / len(flips)
    probs.append(prob_h)
```

```python
print(freqs)
```

```python
plt.plot(probs)
plt.ylabel('Estimated Probability')
plt.xlabel('Number of Flips in Exxperiment');
plt.title("Calculating Probability over Number of Flips")
sns.despine(offset=10, trim=True);  # move axes away 
plt.show()
```
